```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
#the purpose Ui to analayse the data. its mean single variable.
#Data_type 1-->Numerical data(like number) 2----->  categorical Data (gender,nationality etc.)
df=sns.load_dataset("titanic")
df
```

 #1:Categorical_Data (graph the data)

```python
sns.countplot(data=df,x="survived")
df["survived"].value_counts().plot(kind="bar")
```

```python
sns.countplot(data=df,x="pclass",color="red")
```

```python
df["survived"].value_counts().plot(kind="pie",autopct='%.1f%%') #categorical 
```

```python
sns.histplot(data=df,x="age",bins=10)  #for numeric data
df["age"].mean()
```

```python
sns.distplot(df["age"]) #it tell us about the probablity
```

```python
#Box_plot
sns.boxplot(data=df,x="fare")  # max=q3+1.5IQR and min=Q1-1.5*IQR ---->to elimite the outlier its used when data is noisy.it tell us about the data have skewness or not
```

```python
df["age"].skew()*100          #----->to check the skewness of the columns
```

```python
tips=sns.load_dataset("tips")
tips
```

```python
#to analyze the Total sex
sns.countplot(data=tips,x="sex")
tips["sex"].value_counts()
```

```python
sns.countplot(data=tips,x="smoker")
tips["smoker"].value_counts()
```

```python
sns.countplot(data=tips,x="time")
tips["time"].value_counts()
```

```python
sns.countplot(data=tips,x="size")
tips["size"].value_counts()
```

```python
tips["smoker"].value_counts().plot(kind="pie",autopct="%.2f%%")
print("Total_Smoker")
```

```python
sns.histplot(data=tips,x="total_bill")
print("minimum bill in data:",tips["total_bill"].min())
print("maximum bill in data:",tips["total_bill"].max())
print("mean bill in data:",tips["total_bill"].mean())
print("median bill in data:",tips["total_bill"].median())
print("mode bill in data:",tips["total_bill"].mode())
```

```python
sns.distplot(tips["total_bill"])
```

```python
sns.histplot(data=tips,x="tip")
print("minimum tip in bill:",tips["tip"].min())
print("maximum tip in bill:",tips["tip"].max())
print("mean of tip in bill:",tips["tip"].mean())
print("median of tip in bill:",tips["tip"].median())
print("mode of tip in bill:",tips["tip"].mode())
```

```python
sns.distplot(tips["tip"])
```
