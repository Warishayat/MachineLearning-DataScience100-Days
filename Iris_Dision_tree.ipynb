```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.tree import DecisionTreeClassifier
import scipy.stats as stats
from sklearn.preprocessing import PowerTransformer,LabelEncoder
from sklearn.metrics import accuracy_score
df=sns.load_dataset("iris")
df["species"].value_counts()
#now Perform the eda.
df.sample(5)
#check the shape of dataset
df.shape
#to check the null value of data
df.isnull().sum()
#to check the which colum is numerical and which is object
df.info()
#now plot kdeplot of data
sns.kdeplot(data=df)
plt.subplot(121)
sns.boxplot(data=df["sepal_length"])
plt.title("sepal_length")
plt.show()

plt.subplot(122)
sns.boxplot(data=df["sepal_width"])
plt.title("sepal_width")
plt.show()

plt.subplot(121)
sns.boxplot(data=df["petal_length"])
plt.title("petal_length")
plt.show()

plt.subplot(122)
sns.boxplot(data=df["petal_width"])
plt.title("petal_width")
plt.show()

#Now its time to label encode the column species which is not in numeric form it is resultant colum so we will perfoam labelEncoder on that..
Lencode=LabelEncoder()
df["species"]=Lencode.fit_transform(df[["species"]])
#to check the sanple of dataset labelEncode perfoam sucessfully or not
df.sample(5)
#its time to train test split the data
x=df[["sepal_length","sepal_width","petal_length","petal_width"]]
y=df[["species"]]
#now train test the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
#apply the model on the data and check their accuracy.
#import model Disicion Tree classifier
tree=DecisionTreeClassifier() #first make object of model
#fit the data into the model
tree.fit(x_train,y_train)
#check the accuracy of model
y_pred=tree.predict(x_test)
Model_accuracy=accuracy_score(y_test,y_pred)
print("Model have the accuracy of:",Model_accuracy)

#now cross validation if we do it for 10 time then what will be the output
np.mean(cross_val_score(tree,x_train,y_train,scoring="r2",cv=10))*100
#now plot the data before transforming the data.
#iterate the loop on coloum and make graph of each column.
for col in x_train.columns:

    plt.figure(figsize=(12,4))
    plt.subplot(121)
    sns.distplot(df[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(df[col],dist="norm",plot=plt)
    plt.title(col)

#now thesame thing with applying power transformer function.
ptr=PowerTransformer()
#transform the data with the function
x_train_transformed=ptr.fit_transform(x_train)
x_tets_transformed=ptr.fit_transform(x_test)
#now test the model with this colum transform data.
#import model
tree_1=DecisionTreeClassifier()
#fit the data into model
tree_1.fit(x_train_transformed,y_train)
#again check the accuracy of transform data 
accuracy=tree_1.predict(x_tets_transformed)
accuracy_score(y_test,accuracy)

#now i will do the cross validartion of transformed_colum data 10 time
np.mean(cross_val_score(tree_1,x_train_transformed,y_train,scoring="r2",cv=10))*100

#again convert the x_train_transformed into the PandsDataframe
x_train=pd.DataFrame(x_train_transformed,columns=x_train.columns)
#iterate the loop on the transform columns and plot graph of each columns
for col in x_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train[col])
    plt.title(col)
    
    plt.subplot(122)
    stats.probplot(x_train[col],dist="norm",plot=plt)
    plt.title(col)
    

# now take the user input and predict the answer.
sepal_length=float(input("Enter the sepal_length"))
sepal_width=float(input("Enter the sepal_width"))
petal_length=float(input("Enter the petal_length"))
petal_width=float(input("Enter the petal_width"))

user_input=np.array([sepal_length,sepal_width,petal_length,petal_width])
result=tree.predict([user_input])
if result[0]==1:
    print("Species like to setosa")
elif result[0]==0:
    print("Species like to versicolor")
elif result[0]==2:
    print("Species like to virginica")
else:
    print("Outliers")
```
