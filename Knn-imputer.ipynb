```python
import pandas as pd 
import numpy as np
import seaborn as sns
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer,SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=sns.load_dataset("titanic")[["age","fare","embarked","survived"]]
df.head(3)
```

```python
#WE USE KNN IMPUTER BY USNG THAT BY USING THAT WE FILL  THE VALUE 
```

```python
x=df[["age","fare"]]
y=df[["survived"]]
```

```python
x.isnull().mean()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=51)
```

```python
imputer=KNNImputer(n_neighbors=2)
```

```python
x_train_trans=imputer.fit_transform(x_train)
x_test_transf=imputer.fit_transform(x_test)
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train_trans,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=lr.predict(x_test_transf)
```

```python
#check accuracy score with knn
accuracy_score(y_pred,y_test)
```
