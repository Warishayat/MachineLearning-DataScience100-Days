```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#inbuilt_dataset
tip=sns.load_dataset("tips")
titanic=sns.load_dataset("titanic")
flight=sns.load_dataset("flights")
iris=sns.load_dataset("iris")
```

```python
#1:Scatter_plot (Numerical_Numerical)----->Bivariate
#2: Numerical vs categorical
#3: Categorica vs Categorical
```

```python
tip.sample(3)
```

```python
#using scatter plot we willl find rltn between total_bill and tips
plt.figure(figsize=(5,4))
sns.scatterplot(data=tip,x="total_bill",y="tip",hue="sex",style="smoker")  #3multivariate_analysis
```

```python
#Bar_plot(Numerical vs categorical)
titanic.sample(4)
```

```python
sns.barplot(data=titanic,x="pclass",y="fare",hue="who")
```

```python
#Box_plot(Numerical-Categorical)
sns.boxplot(data=titanic,x="sex",y="age")
```

```python
#Distplot(Numerical-Categorical)
sns.distplot(titanic[titanic["survived"]==0]["age"],hist=False)
sns.distplot(titanic[titanic["survived"]==1]["age"],hist=False)
```

```python
#Heatmap (Categorical vs categorical)
titanic.sample(3)

sns.heatmap(pd.crosstab(titanic["pclass"],titanic["survived"]))
```

```python

sns.heatmap(pd.crosstab(titanic["survived"],titanic["sex"]) )
```

```python
sns.clustermap(pd.crosstab(titanic["sibsp"],titanic["survived"]))
```

```python
# pairplot(Numerical vs Numerical)
iris.sample(3)
sns.pairplot(iris)
```
