```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
from sklearn.model_selection import train_test_split,cross_val_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder,PowerTransformer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score,r2_score
```

```python
df=pd.read_csv("concrete_data.csv")
df.head(2)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
x=df[["Cement","Blast Furnace Slag","Fly Ash","Water","Superplasticizer","Coarse Aggregate","Fine Aggregate","Age"]]
y=df[["Strength"]]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
#cross validation score
np.mean(cross_val_score(lr,x,y,scoring="r2"))
```

```python
for col in x_train.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)

    sns.distplot(x_train[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(x_train[col],dist="norm",plot=plt)
    plt.title(col)

    plt.show()
```

```python
#Now with the transformation
```

```python
pt=PowerTransformer()
```

```python
x_train_transformed=pt.fit_transform(x)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
lr=LinearRegression()
np.mean(cross_val_score(lr,x_train_transformed,y,scoring="r2"))
```

```python
x_train_transformed=pd.DataFrame(x_train_transformed,columns=x_train.columns)
```

```python
#Now scale the data.
for col in x_train_transformed.columns:
    plt.figure(figsize=(14,4))
    plt.subplot(121)
    sns.distplot(x_train_transformed[col])
    plt.title(col)

    plt.subplot(122)
    stats.probplot(x_train_transformed[col],dist="norm",plot=plt)
    plt.title(col)

    plt.show()
    
```
